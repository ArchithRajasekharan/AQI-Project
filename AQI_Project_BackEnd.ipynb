```python
import pandas as pd
import numpy as np
data=pd.read_csv(r"/content/sample_data/Final!!.csv")
data
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
data.dropna(inplace=True)
```

```python
data.isnull().sum()
```

```python
data
```

```python
data.columns
```

```python
x=data[[ 'SO2', 'CO', 'NO', 'NO2', 'NOX', 'NH3', 'O3', 'WS',
       'WD', 'RH', 'SR', 'TC']]
y=data['AQI']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
x_train.shape
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(x_train,y_train)
y_pred=reg.predict(x_test)
y_pred
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

```python
from sklearn.neighbors import KNeighborsRegressor
reg1=KNeighborsRegressor()
reg1.fit(x_train,y_train)
y_pred=reg1.predict(x_test)
y_pred
```

```python
r2_score(y_test,y_pred)
```

```python
from sklearn.tree import DecisionTreeRegressor
reg3=DecisionTreeRegressor()
reg3.fit(x_train,y_train)
y_pred=reg3.predict(x_test)
y_pred
```

```python
r2_score(y_test,y_pred)
```

```python
from sklearn.ensemble import RandomForestRegressor
reg4=RandomForestRegressor(n_estimators=100)
reg4.fit(x_train,y_train)
y_pred=reg4.predict(x_test)
y_pred
```

```python
r2_score(y_test,y_pred)
```

```python
import pickle
with open('model.pkl','wb') as file:
  pickle.dump(reg4,file)
```
